```python
# CLEANING THE DATA
df = df.dropna(how="all")
```

```python
df.columns = df.columns.str.strip()
```

```python
cols = ["Order ID","Order Date","Ship Date","Ship Mode","Customer ID","Customer Name","Segment",
        "Country","City","State","Region","Product ID","Category","Sub-Category","Product Name","Payment Mode"]
```

```python
for col in cols:
 df[col] = df[col].str.strip()
```

```python
df.drop(columns=["ind1","ind2"],inplace =True)
```

```python
df["Returns"] = df["Returns"].fillna(0)
```

```python
dates = ["Order Date","Ship Date"]
for date in dates:
    df[date] =  pd.to_datetime(df[date],errors = "coerce")
    df[date] = df[date].dt.strftime("%Y-%m-%d")
    df[date] =  pd.to_datetime(df[date],errors = "coerce")
```

```python
df.rename(columns = {"Row ID+O6G3A1:R6":"Row ID"},inplace= True)
```

```python
df.drop(columns=["Row ID"],inplace=True)
```

```python
df.insert(0,"Row ID",range(1,len(df) +1))
```

```python
df.columns = df.columns.str.replace(" ","_")
```

```python
df.rename(columns = {"sub-category":"sub_category"},inplace = True)
df
```
